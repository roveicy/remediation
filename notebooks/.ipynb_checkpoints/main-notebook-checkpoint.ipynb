### 1 Panel Data Collection
In this section, a set of helper functions is created to collect metric data from Prometheus and injection status from Kubernetes on which Chaos Mesh injection occurs. A time series panel data is created that will be later used for training and testing data sets.

```python
#Import Prometheus API Client
from prometheus_api_client import PrometheusConnect, MetricSnapshotDataFrame, MetricRangeDataFrame
from prometheus_api_client.utils import parse_datetime

#Import Kubernetes API Client
import kubernetes

import pandas as pd
import sktime
import json
from time import sleep
```

```python
#Suppress warnings
import warnings
warnings.filterwarnings('ignore')

#Use this to get warnings back
#warnings.filterwarnings('default')
```

```python
def get_metric_data(start_time, end_time, query, resolution, max_length):
    '''
    Retrieves prometheus metric data and returns a pd.Series
    
        Parameters:
                start_time (string): x start time in seconds (s), minutes (m), hours (h), or days (d) 
                end_time (string): x end time in seconsd (s), minutes (m), hours (h), days (d), or now
                query (string): a PromQL query
                resolution (float): time interval width in seconds, e.g., 1.00 = 1 second width
                max_length (int): will ensure returned data is of fixed-length
        Returns:
                metric_data (pandas.Series): A pandas Series object containing the time series

        Usage: 
            get_metric_data("5m", "now", "rate(container_cpu_usage_seconds_total[1m])", 1.00
    '''

    #Specify time interval
    datetime_start = parse_datetime(start_time)
    datetime_end = parse_datetime(end_time)

    #Query prometheus 
    metric_df = MetricRangeDataFrame(prom.custom_query_range(query, datetime_start, datetime_end, 1.00))
    return metric_df["value"].reset_index().drop("timestamp",axis=1).T.squeeze().tail(max_length)

def get_injection_status(chaos_type, namespace, experiment_name):
    '''
    Retrieves the status of a Chaos Mesh experiment and returns 1 if injection is in progress or 0 if not.
    
        Parameters:
            chaos_type (string): the type of Chaos Mesh injection (e.g., stresschaos)
            namespace (string): the Kubernetes namespace where the experiment was being conducted
            experiment_name: the name of the Chaos Mesh experiment

        Returns:
            injection_status (int): 1 if injection in progress, 0 if injection is paused or completed

        Usage:
            get_injection_status("stresschaos", "wordpress", "my-experiment-1")

    '''

    with kubernetes.client.ApiClient(configuration) as api_client:
        api_instance = kubernetes.client.CustomObjectsApi(api_client)
        group = "chaos-mesh.org"
        version = "v1alpha1"
        plural = "stresschaos"
        name = experiment_name
        namespace = "wordpress"


        api_response = api_instance.get_namespaced_custom_object(group, version, namespace, plural, name)
        experiment_status =  api_response['status']['experiment']['containerRecords'][0]['phase']
        return 1 if experiment_status == 'Injected' else 0

def get_labelled_panel_metric_data(start_time, end_time, query, resolution, max_length, duration, interval, 
                                    chaos_type, namespace, experiment_name):
    '''
    Retrieves Prometheus metric data and returns a "nested_univ" time series panel. Current version supports univariate (single) time series panel.
    
        Parameters:
                start_time (string): x start time in seconds (s), minutes (m), hours (h), or days (d) 
                end_time (string): x end time in seconsd (s), minutes (m), hours (h), days (d), or now
                query (string): a PromQL query
                resolution (float): time interval width in seconds, e.g., 1.00 = 1 second width
                max_length (int): will ensure returned data is of fixed-length
                instances
                duration (int): time in seconds to collect data
                interval (int): the time in seconds to collect between instances
                chaos_type (string): the type of Chaos Mesh injection (e.g., stresschaos)
                namespace (string): the Kubernetes namespace where the experiment was being conducted
                experiment_name: the name of the Chaos Mesh experiment
        Returns:
                panel_data ("nested_univ"): A pandas.DataFrame time series panel conforming to "nested_univ" sktime mtype
                y_list (pd.Series) : a pandas.Series of class labels

        Usage:
                x, y = get_labelled_panel_metric_data(...)
    '''

    x_list = []
    y_list = []

    for i in range(duration):
        x_list.append(get_metric_data(start_time, end_time, query, resolution, max_length))
        y_list.append(get_injection_status(chaos_type, namespace, experiment_name))
        sleep(interval)

    panel_data = pd.DataFrame({'dim_0': x_list})
    
    return panel_data, pd.Series(y_list)
```
